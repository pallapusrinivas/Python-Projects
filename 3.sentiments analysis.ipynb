```python
#1. What is the overall distribution of sentiments (positive, negative, neutral) across the posts?

sentiment_distribution = data['Sentiment'].value_counts(4)       #sentiment is case sensative so always check data properly

print("Sentiment Distribution:")
print(sentiment_distribution)
```

```python
#2. Which platform generates the most engagement in terms of likes and retweets?

data['Total_Engagement'] = data['Likes'] + data['Retweets']
platform_engagement = data.groupby('Platform')['Total_Engagement'].sum().sort_values(ascending=False)

print("Platform Engagement (Total Likes + Retweets):")
print(platform_engagement)
```

```python
#3.Which users have the highest average engagement (likes and retweets per post)?

data['Engagement_Score'] = data['Likes'] + data['Retweets']
user_avg_engagement = data.groupby('User')['Engagement_Score'].mean().sort_values(ascending=False)

print("User Engagement (Average Likes +Average Retweets per Post):")
print(user_avg_engagement.head(10))

```

```python
user_avg_engagement_df = user_avg_engagement.reset_index()
plt.figure(figsize=(20, 5))
sns.barplot(data=user_avg_engagement_df.head(10), x='User', y='Engagement_Score', dodge=True)
plt.xticks(rotation=90)
plt.title('User Engagement (Average Likes +Average Retweets per Post)')
plt.ylabel('Average Engagement Score')
plt.xlabel('User')
plt.show()
```

```python
#To get the last 10 rows.
data['Engagement_Score'] = data['Likes'] + data['Retweets']
user_avg_engagement = data.groupby('User')['Engagement_Score'].mean().sort_values(ascending=False)

print("User Engagement (Average Likes +Average Retweets per Post):")
print(user_avg_engagement.tail(10))
```

```python
user_avg_engagement_df = user_avg_engagement.reset_index()
plt.figure(figsize=(10, 5))
sns.barplot(data=user_avg_engagement_df.tail(10), x='User', y='Engagement_Score', dodge=True)
plt.xticks(rotation=90)
plt.title('User Engagement (Average Likes +Average Retweets per Post)')
plt.ylabel('Average Engagement Score')
plt.xlabel('User')
plt.show()
```
