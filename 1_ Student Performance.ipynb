```python
data = data.isna()
```

```python
categorical_columns = ['School_Type', 'Parental_Education_Level']
data[categorical_columns] = data[categorical_columns].astype('category')
```

```python
data.info()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['Exam_Score'], kde=True, bins=20, color='blue')
plt.title('Distribution of  Exam Scores')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.show()
```

```python
#1. How does the number of hours studied per week correlate with the final exam score?

plt.figure(figsize=(8, 6))
sns.scatterplot(x=data['Hours_Studied'], y=data['Exam_Score'])
plt.title('Hours Studied vs. Exam Score')
plt.xlabel('Hours Studied ')
plt.ylabel(' Exam Score')
plt.show()
```

```python
# correlation coefficient
corr_attendance, _ = pearsonr(data['Attendance'], data['Exam_Score'])
print(f"Correlation between attendance and exam score: {corr_attendance:.2f}")

```

```python
#2.Is there a significant relationship between class attendance percentage and exam scores?

plt.figure(figsize=(20, 10))
sns.barplot(x=data['Attendance'], y=data['Exam_Score'])
plt.title('Attendance vs. Exam Score')
plt.xlabel('Attendance')
plt.ylabel('Exam Score')
plt.show()
```

```python
#significant
corr_attendance, _ = pearsonr(data['Attendance'], data['Exam_Score'])
print(f"Correlation between attendance and exam score: {corr_attendance:.2f}")

```

```python
#3. Does the average number of sleep hours per night influence exam performance?

plt.figure(figsize=(8, 6))
sns.lineplot(x=data['Sleep_Hours'], y=data['Exam_Score'])
plt.title('Sleep Hours vs. Final Exam Score')
plt.xlabel('Sleep Hours')
plt.ylabel(' Exam Score')
plt.show()
```

```python
#Correlation coefficient

corr_sleep, _ = pearsonr(data['Sleep_Hours'], data['Exam_Score'])
print(f"Correlation between sleep hours and exam score: {corr_sleep:.2f}")
```

```python
#4.How does family income and parental education level impact the student's exam scores?

plt.figure(figsize=(10, 5))
sns.boxplot(x='Family_Income', y='Exam_Score', data=data)
plt.title('Family Income  vs.  Exam Score')
plt.xlabel('Family Income ')
plt.ylabel(' Exam Score')
plt.show()
```

```python
#Family Income

anova_income = f_oneway(*[data[data['Family_Income'] == level]['Exam_Score']
                          for level in data['Family_Income'].unique()])
print(f"ANOVA result for Family Income Levels: F-statistic={anova_income.statistic:.2f}, p-value={anova_income.pvalue:.2e}")
```

```python
# Parental Education.

plt.figure(figsize=(8, 6))
sns.boxplot(x='Parental_Education_Level', y='Exam_Score', data=data)
plt.title('Parental Education Level vs. Final Exam Score')
plt.xlabel('Parental Education Level')
plt.ylabel('Final Exam Score')
plt.show()
```

```python
# Parental Education Level

anova_education = f_oneway(*[data[data['Parental_Education_Level'] == level]['Exam_Score']
                             for level in data['Parental_Education_Level'].unique()])
print(f"ANOVA result for Parental Education Levels: F-statistic={anova_education.statistic:.2f}, p-value={anova_education.pvalue:.2e}")
```

```python
#5. Do private school students perform better in exams than public school students?

plt.figure(figsize=(10, 6))
sns.histplot(x='School_Type', y='Exam_Score', data=data)
plt.title('School Type vs. Final Exam Score')
plt.xlabel('School Type')
plt.ylabel('Final Exam Score')
plt.show()
```

```python
# School Type

private_scores = data[data['School_Type'] == 'Private']['Exam_Score']
public_scores = data[data['School_Type'] == 'Public']['Exam_Score']
t_stat, p_value = ttest_ind(private_scores, public_scores)
print(f"T-test result: t-statistic={t_stat:.2f}, p-value={p_value:.2e}")
```
