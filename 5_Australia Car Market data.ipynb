```python
#1.What is the average price of cars across different brands, and which brands have the highest and lowest average prices?

# Calculate average price by brand
avg_price_by_brand = df.groupby('Brand')['Price'].mean().reset_index()

# Find brands with the highest and lowest average prices
highest_avg_price = avg_price_by_brand.sort_values('Price', ascending=False).head(5)
lowest_avg_price = avg_price_by_brand.sort_values('Price', ascending=True).head(5)

print("Brands with the highest average prices:")
print(highest_avg_price)

print("\nBrands with the lowest average prices:")
print(lowest_avg_price)
```

```python
#2. How does the condition of a car (Status) affect its price?

import seaborn as sns
import matplotlib.pyplot as plt

# Group by Status and calculate average price
avg_price_by_status = df.groupby('Status')['Price'].mean().reset_index()

print("Average price by car condition (Status):")
print(avg_price_by_status)

# Visualization
plt.figure(figsize=(15, 10))
sns.boxplot(data=df, x='Status', y='Price')
plt.title('Car Price by Condition (Status)')
plt.xlabel('Condition (Status)')
plt.ylabel('Price')
plt.show()
```

```python
#3.Is there a correlation between mileage (Kilometers) and the price of a car?

# Check correlation
correlation = df['Kilometers'].corr(df['Price'])
print(f"Correlation between Kilometers and Price: {correlation}")

# Scatter plot
sns.lineplot(data=df, x='Kilometers', y='Price', alpha=0.5)
plt.title('Mileage vs Car Price')
plt.xlabel('Kilometers')
plt.ylabel('Price')
plt.show()

```

```python
#4.How does engine capacity (CC) relate to car price and mileage?

# Correlation analysis
correlation_price_cc = df['CC'].corr(df['Price'])
correlation_mileage_cc = df['CC'].corr(df['Kilometers'])

print(f"Correlation between Engine Capacity (CC) and Price: {correlation_price_cc}")
print(f"Correlation between Engine Capacity (CC) and Mileage: {correlation_mileage_cc}")

# Scatter plots
plt.figure(figsize=(12, 10))

# CC vs Price
plt.subplot(1, 2, 1)
sns.barplot(data=df, x='CC', y='Price', alpha=0.5)
plt.title('Engine Capacity (CC) vs Price')
plt.xlabel('Engine Capacity (CC)')
plt.ylabel('Price')

# CC vs Mileage
plt.subplot(1, 2, 2)
sns.barplot(data=df, x='CC', y='Kilometers', alpha=0.5)
plt.title('Engine Capacity (CC) vs Mileage')
plt.xlabel('Engine Capacity (CC)')
plt.ylabel('Kilometers')

plt.tight_layout()
plt.show()
```
